```python
class DoublyNode:
    def __init__(self, value):
        self.value      = value
        self.next       = None
        self.previous   = None

class CircularDoublyLinkedList:
    def __init__(self):
        self.head = None
        self.tail = None

    def __iter__(self):
        iterNode = self.head
        while iterNode:
            yield iterNode
            if iterNode.next == self.head:
                break
            iterNode = iterNode.next

    def create(self, value):
        node = DoublyNode(value)
        self.head = node
        self.tail = node
        node.previous = node
        node.next = node
    
    def insert(self, value, location):
        # Time complexity: O(n)
        if self.head == None:
            return "Circular Doubly Linked List is empty"
        else:
            newNode = DoublyNode(value)
            if location == 0:                   # Insert at first position 
                self.head.previous = newNode
                self.tail.next = newNode
                newNode.next = self.head
                newNode.previous = self.tail
                self.head = newNode
            elif location == -1:                # Insert at last position
                self.tail.next = newNode
                self.head.previous = newNode
                newNode.next = self.head
                newNode.previous = self.tail
                self.tail = newNode
            else:                               # Insert at middle position
                index = 0
                loopNode = self.head
                while index < location - 1:
                    index += 1
                    loopNode = loopNode.next
                loopNode.next.previous = newNode
                newNode.next = loopNode.next
                loopNode.next = newNode
                newNode.previous = loopNode

    def traversal(self):
        iterNode = self.head
        while iterNode:
            print(iterNode.value)
            if iterNode.next == self.head:
                break
            iterNode = iterNode.next

    def reverseTraversal(self):
        iterNode = self.tail
        while iterNode:
            print(iterNode.value)
            if iterNode.previous == self.tail:
                break
            iterNode = iterNode.previous

    def search(self, value):
        iterNode = self.head
        index = 0
        while iterNode:
            if iterNode.value == value:
                return "Found value: {} at index: {}".format(value, index)
            if iterNode.next == self.head:
                return "Value {} does'nt exist in Linked List".format(value)
            iterNode = iterNode.next
            index += 1

    def delete(self, location):
        if self.head == None:
            return "Doubly Linked List is Empty"
        else:
            if location == 0:           # Delete first node in linked list
                self.tail.next = self.head.next
                self.head.next.previous = self.tail
                self.head.next = None
                self.head.previous = None
                self.head = self.tail.next
            elif location == -1:        # Delete last node in linked list
                node = self.tail.previous
                self.head.previous = node
                node.next = self.head
                self.tail.next = None
                self.tail.previous = None
                self.tail = node
            else:
                index = 0
                iterNode = self.head
                while index < location:
                    index += 1
                    iterNode = iterNode.next
                node = iterNode.previous
                node.next = iterNode.next
                iterNode.next.previous = node
                iterNode.next = None
                iterNode.previous = None

    def deleteEntire(self):
        self.tail.next = None
        self.head.previous = None
        iterNode = self.head
        while iterNode.next != None:
            iterNode.previous = None
            iterNode = iterNode.next
        self.tail = None
        self.head = None
```

```python
a = CircularDoublyLinkedList()
a.create(23)
a.insert(450, 0)
a.insert(12, -1)
a.insert(34, -1)
a.insert(45, 0)
a.insert(1, 2)
print([x.value for x in a])
# a.delete(0)
# a.delete(-1)
a.delete(2)
a.deleteEntire()
print([x.value for x in a])
```
