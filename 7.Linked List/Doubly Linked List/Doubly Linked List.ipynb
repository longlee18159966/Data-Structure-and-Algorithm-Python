```python
class NodeDoubly:
    def __init__(self, value):
        self.value      = value
        self.prev       = None
        self.next       = None

class DoublyLinkedList:
    def __init__(self):
        self.tail = None
        self.head = None
    
    def create(self, value):
        newNode = NodeDoubly(value)
        self.tail = newNode
        self.head = newNode
        return "Doubly Linked List is created"

    def __iter__(self):
        tempNode = self.head
        while tempNode:
            yield tempNode
            tempNode = tempNode.next

    def insert(self, value, location):
        # Time complexity: O(n)
        if self.head == None:
            return "Doubly Linked List is empty"
        else:
            newNode = NodeDoubly(value)
            if location == 0:                       # Insert at first position
                newNode.next = self.head
                self.head.prev = newNode
                self.head = newNode
            elif location == -1:                      # Insert at last position
                self.tail.next = newNode
                newNode.prev = self.tail
                self.tail = newNode
            else:                                   # Insert at middle position
                index = 0
                temp = self.head
                while index < location - 1:
                    index += 1
                    temp = temp.next
                newNode.next        = temp.next
                newNode.prev        = temp
                newNode.next.prev   = newNode
                temp.next           = newNode
                if temp == self.tail:
                    self.tail = newNode

    def traversal(self):
        iterNode = self.head
        while iterNode:
            print(iterNode.value)
            iterNode = iterNode.next

    def traversalReverse(self):
        iterNode = self.tail
        while iterNode:
            print(iterNode.value)
            iterNode = iterNode.prev

    def search(self, value):
        if self.head == None:
            return "Doubly Linked List is empty"
        iterNode = self.head
        index= 0
        while iterNode:
            if iterNode.value == value:
                return "Found value {} at index {}".format(value, index)
            index += 1
            iterNode = iterNode.next
        return "Value {} doesn't exist in Doubly Linked List".format(value)

    def delete(self, location):
        if self.head == None:
            return "Doubly Linked List is empty"
        else:
            if location == 0:                           # Delete first position
                if self.head.next == None:              # Linked List has one element
                    self.head = None
                    self.tail = None
                else:
                    self.head = self.head.next
                    self.head.prev = None
            elif location == -1:                        # Delete last position
                if self.head.next == None:              # Linked List has one element
                    self.head = None
                    self.tail = None
                else:
                    self.tail = self.tail.prev
                    self.tail.next.prev = None
                    self.tail.next = None
            else:                                       # Delete middle position
                if self.head.next == None:              
                    self.head = None
                    self.tail = None
                else:
                    index = 0
                    iterNode = self.head
                    while index < location:
                        index += 1
                        iterNode = iterNode.next
                    iterNode.prev.next = iterNode.next
                    iterNode.next.prev = iterNode.next

    def deleteEntire(self):
        temp = self.head
        while temp:
            temp.prev = None
            temp = temp.next
        self.head = None
        self.tail = None


```
