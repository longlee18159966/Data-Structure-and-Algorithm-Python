```python
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None

class CircularLinkedList:
    def __init__(self):
        self.head = None
        self.tail = None
    
    def __iter__(self):
        node = self.head
        while node:
            yield node
            if node.next == self.head:
                break
            node = node.next

    def create(self, value):
        newNode = Node(value)
        self.head = newNode
        self.tail = newNode
        self.tail.next = self.head
        return "New Cirular Linked List has been created"

    def insert(self, value, location):
        # Time complexity: O(n)
        if self.head == None:                               # CSLL is empty
            self.create(value)
        else:                                                                   
            newNode = Node(value)
            if location == 0:                               # Insert at beginning
                newNode.next = self.head
                self.head = newNode
                self.tail.next = self.head
            elif location == -1:                            # Insert at last
                newNode.next = self.tail.next
                self.tail.next = newNode
                self.tail = newNode
            else:                                           # Insert at middle
                index = 0
                temp = self.head
                while index < location - 1:
                    index += 1
                    temp = temp.next
                nextNode = temp.next
                temp.next = newNode
                newNode.next = nextNode
                if temp == self.tail:
                    self.tail = newNode

    def traversal(self):
        # Time complexity: O(n)
        if self.head == None:
            return "Circular Linked List is empty"
        iterNode = self.head
        while iterNode:
            print(iterNode.value)
            iterNode = iterNode.next
            if iterNode == self.head:
                break

    def search(self, value):
        # Time complexity: O(n)
        if self.head == None:
            return "Circular Linked List is empty"
        else:
            index = 0
            temp = self.head
            while temp:
                if temp.value == value:
                    return "Found at index {}".format(index)
                index += 1
                temp = temp.next
                if temp == self.tail.next:
                    return "Did'nt find value {} in Circular Linked List".format(value)

    def delete(self, location):
        # Time complexity: O(n)
        if self.head == None:
            return "Circular Linked List is empty"
        else:
            if location == 0:                   # Delete first location
                if self.head == self.tail:      # Linked List has one node
                    self.head.next = None
                    self.head = None
                    self.tail = None
                else:
                    headNode = self.head.next
                    self.head = headNode
                    self.tail.next = headNode
            elif location == -1:                # Delete last location
                if self.head == self.tail:      # Linked List has one node
                    self.head.next = None
                    self.head = None
                    self.tail = None
                else:
                    tempNode = self.head
                    while tempNode.next.next != self.head:
                        tempNode = tempNode.next
                    tempNode.next = self.head
                    self.tail = tempNode
            else:                               # Delete middle location
                if self.head == self.tail:      # Linked List has one node
                    self.head.next = None
                    self.head = None
                    self.tail = None
                else:
                    index = 0
                    tempNode = self.head
                    while index < location - 1:
                        index += 1
                        tempNode = tempNode.next
                    tempNode.next = tempNode.next.next

    def deleteEntire(self):
        self.head = None
        self.tail.next = None
        self.tail = None
```

```python
csll = CircularLinkedList()
csll.create(12)
csll.insert(14, 0)
csll.insert(23, 0)
csll.insert(123, -1)
csll.insert(1,2)
print([i.value for i in csll])
csll.deleteEntire()
print([i.value for i in csll])
```
