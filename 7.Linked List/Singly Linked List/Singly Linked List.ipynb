```python
class Node:
    def __init__(self, value=None):
        self.value = value
        self.next = None
```

```python
class SinglyLinkedList:
    def __init__(self):
        self.head = None
        self.tail = None
```

```python
SLL = SinglyLinkedList()
node1 = Node(1)
node2 = Node(2)
```

```python
# Time complexity: O(1)
# Space complexity: O(n)
SLL.head = node1
SLL.head.next = node2
SLL.tail = node2
```

## Delete node from Singly Linked List

```python
# Algorithm
from IPython.display import Image
Image(filename="algorithms_delete.PNG")
```

```python
class SinglyLinkedList:
    def __init__(self):
        self.head = None
        self.tail = None
    
    def __iter__(self):
        # Time complexity: O(n)
        # Space complexity: O(1)
        node = self.head
        while node:
            yield node
            node = node.next

    def insertSLL(self, value, location):
        newNode = Node(value)
        if self.head == None:
            self.head = newNode
            self.tail = newNode
        else:
            # Insert at first location
            if location == 0:
                newNode.next = self.head
                self.head = newNode
            else:
                # Insert at last location
                if location == -1:
                    newNode.next = None
                    self.tail.next = newNode
                    self.tail = newNode
                else:
                    # Insert at middle location
                    tempNode = self.head
                    index = 0
                    while index < location - 1:         # Time complexity: O(n)
                        tempNode = tempNode.next
                        index += 1
                    newNode.next = tempNode.next
                    tempNode.next = newNode
                    if tempNode == self.tail:
                        self.tail = newNode

    def traversalSLL(self):
        # Time complexity: O(n)
        # Space complexity: O(1)
        iterNode = self.head
        if iterNode == None:
            return "Singly Linked List does not exist"
        else:
            while iterNode != None: # or we can write: While iterNode is not None:
                print(iterNode.value)
                iterNode = iterNode.next

    def searchSLL(self, searchValue):
        # Time complexity: O(n)
        # Space complexity: O(1)
        iterNode = self.head
        if iterNode == None:
            return "Singly Linked List does not exist"
        else:
            position = 1
            while iterNode != None:
                if iterNode.value == searchValue:
                    return "Found, position: {}(index = {}), value: {}".format(position, position-1, searchValue)
                position += 1
                iterNode = iterNode.next
            return "Value {} does not exist in Linked List".format(searchValue)

    def deleteNodeInSLL(self, location):
        if self.head == None:
            return 'Linked List is empty'
        else:   # Linked List is not empty
            if location == 0:                       # Delete first node
                if self.head.next == None:          # if Linked List has one node
                    self.head = None
                    self.tail = None
                else:                               # Linked List has many node (> 1)
                    self.head = self.head.next
            elif location == -1:                    # Delte last node
                if self.head.next == None:          # if Linked List has one node
                    self.head = None
                    self.tail = None
                else:                               # Linked List has many node (> 1)
                    iterNode = self.head
                    while iterNode != None:
                        if iterNode.next == self.tail:
                            break
                        iterNode = iterNode.next
                    iterNode.next = None
                    self.tail = iterNode
            else:                                   # Delete at middle location
                # Find location
                index = 0
                iterNode = self.head
                while iterNode != None:
                    if index == location-1:
                        break
                    index += 1
                    iterNode = iterNode.next
                # Delete node
                iterNode.next = iterNode.next.next

    def deleteEntireSLL(self):
        if self.head == None:
            return "Singly Linked List is empty"
        else:
            self.head = None
            self.tail = None
```

```python
sll = SinglyLinkedList()
sll.insertSLL(1,0)
sll.insertSLL(13, 0)
sll.insertSLL(34, -1)
sll.insertSLL(0, 0)
sll.insertSLL(15, 2)
print([node.value for node in sll])
sll.searchSLL(34)
sll.deleteNodeInSLL(0)
sll.traversalSLL()
```
