```python
mydict = {'name': 'long', 'age':22, 'address': 'ba vi'}
# Traverse
def traverseDict(dict):
    for key in dict:
        print('{}: {}'.format(key, dict[key]))

traverseDict(mydict)
```

```python
# Search
def searchDict(dict, value):
    for key in dict.keys():
        if dict[key] == value:
            return (key, value)
    return 'not found'
searchDict(mydict, 22)
```

```python
# Delete
print(mydict.pop('age'))
print(mydict.popitem())
del mydict['name']
```
