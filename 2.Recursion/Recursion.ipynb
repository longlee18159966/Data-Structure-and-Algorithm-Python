```python
# Russian Doll
def openRussianDoll(doll):
    if doll == 1:
        print("All dolls are opened")
    else:
        openRussianDoll(doll-1)
openRussianDoll(4)
```

```python
# Find factorial number
def factorial(n):
    assert n >= 0 and int(n) == n, 'The number must be positive integer only'
    if n in [0,1]:
        return 1
    else:
        return n * factorial(n-1)

factorial(4)
```

```python
# Fibonacci numbers
def fibonacci(n):
    assert n >= 0 and int(n) == n, "The number must be positive integer only"
    if n in [0, 1]:
        return n
    else:
        return fibonacci(n-1) + fibonacci(n-2)

fibonacci(20)
```

### Question 1: Sum of Digits

```python
def sumOfDigits(number):
    assert number >= 0 and number == int(number), "The number must be positive integer only"
    number = abs(number)
    if number == 0:
        return number
    return number % 10 + sumOfDigits(number//10)

sumOfDigits(123)
```

### Question 2: Power

```python
def power(number, exponent):
    assert exponent >= 0 and exponent == int(exponent), "The exponent must be positive integer only"
    if exponent == 0:
        return 1
    else:
        return number * power(number, exponent - 1)

power(3.2, 2)
```

### Question 3: Greatest Common Divisor

```python
# Note: Using Euclidean algorithm
def GCD(first, second):
    assert int(first) == first and int(second) == second, "The numbers must be integer only!"
    first = abs(first)
    second = abs(second)
    if second == 0:
        return first
    else:
        return GCD(second, first % second)

GCD(18, 48)
        
```

### Question 4: Decimal To Binary

```python
def decimalToBinary(number):
    if number == 0:
        return 0
    else:
        return  (number % 2 + 10 * decimalToBinary(number // 2))    # Formular 1*
x = 10
decimalToBinary(x)
```
