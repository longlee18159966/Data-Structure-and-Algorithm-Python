```python
import numpy as np
twoDimensionalArray = np.array([[1,2, 3, 4], [4, 5, 6, 7]])
twoDimensionalArray
```

## Insertion - Two Dimensional Array

```python
# Add new column
newTwoDinemsionalArray1 = np.insert(twoDimensionalArray, 0, [2, 4], axis=1)
newTwoDinemsionalArray1
```

```python
# Add new row
newTwoDinemsionalArray2 = np.insert(twoDimensionalArray, 0, [5, 6, 7, 8], axis=0)
newTwoDinemsionalArray2
# Note: axis=1 : columns, axis=0 : rows
```

```python
# Add new row at the end
a = np.append(twoDimensionalArray, [[12, 13, 23, 56]], axis=0)
a
```

## Accessing an element of Two Dimensional Array

```python
def accessElements(arr, rowIndex, columnIndex):
    # Time complexity: O(1)
    # Space complexity: O(1)
    if rowIndex >= len(arr) or columnIndex >= len(arr[0]):
        print("Incorrect index")
    else:
        print("Row: {}, Column: {}, Value: {}".format(rowIndex, columnIndex, arr[rowIndex][columnIndex]))

accessElements(a, 2, 3) 
```

## Traversal

```python
def traversalTDArray(arr):
    # Time complexity: O(m*n)
    # Space complexity: O(1)
    for i in range(len(arr)):
        for j in range(len(arr[0])):
            print(arr[i][j])

traversalTDArray(a)
```

## Searching for an element in Two Dimensional Array

```python
def searchTDArray(arr, value):
    # Time complexity: O(m*n)
    # Space complexity: O(1)
    for i in range(len(arr)):
        for j in range(len(arr[0])):
            if arr[i][j] == value:
                return "Finded value: {} at row: {}, column: {}".format(value, i, j)
    return "Value is not found"
        
searchTDArray(a, 23)
```

## Deletion Two Dimensional Array
